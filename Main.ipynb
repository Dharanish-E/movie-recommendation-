```python
import pandas as pd
import numpy as np
import difflib  # it is used when the user is trying to access the movies via /searching using type and user get it spelled 
                # wrong and to find the similarity between the words we use this !
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv("data.csv")
# movies["title"].to_string()
```

```python
# featuring data
# needed to be filtered [id, title, genre, overview, popularity, vote_count]
movies.columns
```

```python
movies =  movies[['id', 'title', 'overview', 'genre', 'popularity', 'vote_average']]
movies
```

```python
# Combine datas
movies = movies.assign(tags=movies['overview'] + movies['genre'])
movies
```

```python
new_data = movies.drop(columns=['overview', 'genre'])
# new_data.to_string()
```

```python
# textual datas are required to convert into another form called vector --->as it is a representation of numerical values!
# we use Bag of word / TFID
```

```python
# from sklearn.feature_extraction.text import CountVectorizer which converts the textual data to numerical data!
```

```python
cv = CountVectorizer(max_features = 10000, stop_words = 'english')
cv
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
#  using cosine similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
new_data[new_data['title'] == 'Captain America'].index[0]
```

```python
sorted(list(enumerate(similarity[2])), reverse=True)
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True , key=lambda vector:vector[1] )
for i in distance[0:5]:
    print(i)
```

```python
distance = sorted(list(enumerate(similarity[22])), reverse=True , key=lambda vector:vector[1] )
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommand(movi):
    movies = movi.title()
    index = new_data[new_data["title"] == movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse= True, key= lambda vector: vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommand("iron man")
```
